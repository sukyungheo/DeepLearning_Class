### 02. Gradient Descent
<pre>
MSSubClass: The building class
OverallQual: Overall material and finish quality
OverallCond: Overall condition rating
BsmtFinSF1: Type 1 finished square feet
BsmtFinSF2: Type 2 finished square feet
</pre>

```python
tr.info()
```

```python
selCol = ["MSSubClass","LotArea", 
          "OverallQual", "OverallCond", 
          "YearBuilt", "YearRemodAdd",
          "BsmtFinSF1", "BsmtFinSF2", 
          "SalePrice"]
X = tr[selCol]
X.head()
```

```python
# from sklearn.preprocessing import StandardScaler

# scaler = StandardScaler()
# scaler.fit(X)
# X_scaled = scaler.transform(X)
# X_scaled_df = pd.DataFrame(X_scaled, columns=selCol)
# print(type(X_scaled_df))
# print(X_scaled_df.shape, X_scaled_df.head())
```

```python
# MSSubClass, LotFrontage, LotArea, 
x1 = tr["MSSubClass"].values
x2 = tr["OverallQual"].values   
x3 = tr["OverallCond"].values
x4 = tr["BsmtFinSF1"].values
x5 = tr["BsmtFinSF2"].values

w1 = np.random.uniform(low=0.0, high=1.0)
w2 = np.random.uniform(low=0.0, high=1.0)
w3 = np.random.uniform(low=0.0, high=1.0)
w4 = np.random.uniform(low=0.0, high=1.0)
w5 = np.random.uniform(low=0.0, high=1.0)

b = np.random.uniform(low=0.0, high=1.0)

y = tr["SalePrice"].values
```

```python
num_epoch = 10000
learning_rate = 0.00000005

for epoch in range(num_epoch):
    y_predict = x1 * w1 + \
                x2 * w2 + \
                x3 * w3 + \
                x4 * w4 + \
                x5 * w5 + b
                
    error = np.abs(y_predict - y).mean()
    
    w1 = w1 - learning_rate * ((y_predict - y) * x1).mean()
    w2 = w2 - learning_rate * ((y_predict - y) * x2).mean()
    w3 = w3 - learning_rate * ((y_predict - y) * x3).mean()
    w4 = w4 - learning_rate * ((y_predict - y) * x4).mean()
    w5 = w5 - learning_rate * ((y_predict - y) * x5).mean()
    
    b = b - learning_rate * (y_predict - y).mean()
    
    if epoch % 500 ==0:
        print("{0} error={1}".format(epoch, error))
        
print("{0} error={1}".format(epoch, error))
print(w1, w2, w3, w4, w5)
```

```python
result = tr.copy()

y_predict = x1 * w1 + \
            x2 * w2 + \
            x3 * w3 + \
            x4 * w4 + \
            x5 * w5 +  b

result["SalePrice(predict)"] = y_predict
result.head()
```

### 03. 모든 int 변수 컬럼에 대상
<pre>
Gradient descent 알고리즘 아래 항목에 대해 수행해 보자.
</pre>

```python
tr.info()
```

```python
selCol1 = [
"MSSubClass",   
"LotArea", 
"OverallQual", 
"OverallCond", 
"1stFlrSF",
"GrLivArea",
"BsmtFinSF1", 
"BsmtFinSF2", 
"BsmtUnfSF", 
"TotalBsmtSF"    
]
X = tr[selCol1]
print(X.shape)
X.head()
```

```python
num_epoch = 30000
learning_rate = 0.000000000005

w = np.random.uniform(low=0.0, high=1.0, size=10)
b = np.random.uniform(low=0.0, high=1.0)

for epoch in range(num_epoch):
    y_predict = X.dot(w) + b

    error = np.abs(y_predict - y).mean()
    #print(error)
    if error < 5:
        break

    
    w = w - learning_rate * X.T.dot((y_predict - y))
    b = b - learning_rate * (y_predict - y).mean()

    if epoch % 3000 == 0:
        print("{0:4} error = {1:.5f}".format(epoch, error))
        print(w)
        
print("{0:4} error = {1:.5f}".format(epoch, error))
```

```python
result = tr.copy()

y_predict = X.dot(w) + b

result["SalePrice(predict)"] = y_predict
result.head()
```
