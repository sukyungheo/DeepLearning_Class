## 클래스
## 파이썬에서 기본적으로 제공하는 자료형
<pre>
class 클래스이름:
    def __init__(self, 인수, ...):     # 생성자
        ...
    def 메서드이름1(self, 인수, ...):  # 메서드 1
        ...
    def 메서드이름2(self, 인수, ...):  # 메서드 2
        ...
</pre>

```python
class Man:
    def __init__(self, name):
        self.name=name
        print("Initialized!")
        
    def hello(self):
        print("Hello " + self.name + "!")
        
    def goodbye(self):
        print("Good-bye " + self.name + "!")
        
m = Man("David")
m.hello()
m.goodbye()
```

## 1.6 matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# 데이터 준비
x = np.arange(0, 6, 0.1)   # 0에서 6까지 0.1 간격으로 생성
y = np.sin(x)

# 그래프 그리기
plt.plot(x,y)
plt.show()
```
